```python
import matplotlib.pyplot as plt
import numpy as np

data = {
    5: [1.04, 0.79, 0.71, 0.87, 0.83],
    10: [1.10, 1.00, 0.85, 0.76, 0.64],
    15: [0.87, 1.03, 1.19, 1.08, 0.58],
    20: [0.81, 1.46, 2.93, 3.26, 3.49],
    25: [0.95, 1.81, 3.56, 6.45, 10.26],
    30: [0.93, 1.86, 3.72, 6.93, 12.25],
    32: [0.94, 1.87, 3.80, 7.37, 13.40],
    34: [0.96, 1.90, 3.80, 7.52, 13.40]
}
threads = [1, 2, 4, 8, 16]

# Split data into two groups
data_keys = list(data.keys())
mid_index = len(data_keys) // 2
data_first_half = {k: data[k] for k in data_keys[:mid_index]}
data_second_half = {k: data[k] for k in data_keys[mid_index:]}

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

# Plot data for the first half
for d, scalabilities in data_first_half.items():
    ax1.plot(threads, scalabilities, marker='o', label=f'd={d}')
ax1.set_title('Scalability vs. Number of Threads (smaller values of d)', fontsize=16)
ax1.set_xlabel('Number of Threads', fontsize=14)
ax1.set_ylabel('Scalability', fontsize=14)
ax1.set_xticks(threads)
ax1.set_xticklabels([str(t) for t in threads])
ax1.grid(True, which='both', linestyle='--', linewidth=0.5)
ax1.legend(title='Values of d', fontsize=12, title_fontsize=14)

# Plot data for the second half
for d, scalabilities in data_second_half.items():
    ax2.plot(threads, scalabilities, marker='o', label=f'd={d}')
ax2.set_title('Scalability vs. Number of Threads (bigger values of d)', fontsize=16)
ax2.set_xlabel('Number of Threads', fontsize=14)
ax2.set_ylabel('Scalability', fontsize=14)
ax2.set_xticks(threads)
ax2.set_xticklabels([str(t) for t in threads])
ax2.grid(True, which='both', linestyle='--', linewidth=0.5)
ax2.legend(title='Values of d', fontsize=12, title_fontsize=14)

plt.tight_layout()

# Save the plot to a PDF file
plt.savefig('scalability_plot.pdf')

# Display the plot
plt.show()
```
